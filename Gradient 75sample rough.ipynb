```python
class GradientDescent:
    def __init__(self, iteration, learning_rate, train_x, train_y, test_x, test_y):
        self.iter = iteration
        self.learning_rate = learning_rate
        self.w0 = 0
        self.w1 = 0
        self.w2 = 0
        self.w3 = 0
        self.w4 = 0
        self.train_x = train_x
        self.train_y = train_y
        self.test_x = test_x
        self.test_y = test_y

    def cost_find(self):
        cost = 0
        for i in range(len(self.train_x)):
            y_pred = self.w0 + self.w1 * self.train_x.iloc[i, 0] + self.w2 * self.train_x.iloc[i, 1] + self.w3 * self.train_x.iloc[i, 2] + self.w4 * self.train_x.iloc[i, 3]
            cost += (self.train_y.iloc[i, 0] - y_pred) ** 2
        cost /= (2 * len(self.train_x))
        return cost

    def direction(self):
        d = [0] * 5
        for i in range(len(self.train_x)):
            error = self.train_y.iloc[i, 0] - (self.w0 + self.w1 * self.train_x.iloc[i, 0] + self.w2 * self.train_x.iloc[i, 1] + self.w3 * self.train_x.iloc[i, 2] + self.w4 * self.train_x.iloc[i, 3])
            d[0] += error
            d[1] += error * self.train_x.iloc[i, 0]
            d[2] += error * self.train_x.iloc[i, 1]
            d[3] += error * self.train_x.iloc[i, 2]
            d[4] += error * self.train_x.iloc[i, 3]
        return [-lr for lr in d]

    def update_weight(self, d):
        self.w0 -= self.learning_rate * d[0]
        self.w1 -= self.learning_rate * d[1]
        self.w2 -= self.learning_rate * d[2]
        self.w3 -= self.learning_rate * d[3]
        self.w4 -= self.learning_rate * d[4]

    def gradient_descent(self):
        for i in range(self.iter):
            cost = self.cost_find()
            d = self.direction()
            self.update_weight(d)
            print(f"Iteration {i + 1}, Cost: {cost}, Weights: {self.w0}, {self.w1}, {self.w2}, {self.w3}, {self.w4}")


if __name__ == "__main__":
    # Define and initialize train_x, train_y, test_x, test_y
    obj = GradientDescent(500, 0.0001, train_x, train_y, test_x, test_y)
    obj.gradient_descent()
```

```python
import matplotlib.pyplot as plt
import numpy as np

class GradientDescent:
    def __init__(self, iteration, learning_rate, train_x, train_y, test_x, test_y):
        self.iter = iteration
        self.learning_rate = learning_rate
        self.w0 = 0
        self.w1 = 0
        self.w2 = 0
        self.w3 = 0
        self.w4 = 0
        self.train_x = train_x
        self.train_y = train_y
        self.test_x = test_x
        self.test_y = test_y
        self.cost_history = []

    def cost_find(self):
        cost = 0
        for i in range(len(self.train_x)):
            y_pred = self.w0 + self.w1 * self.train_x.iloc[i, 0] + self.w2 * self.train_x.iloc[i, 1] + self.w3 * self.train_x.iloc[i, 2] + self.w4 * self.train_x.iloc[i, 3]
            try:
                cost += np.float64(self.train_y.iloc[i, 0] - y_pred) ** 2
            except OverflowError:
                # Handle overflow gracefully
                print("Overflow encountered in calculation. Adjusting...")
            cost += np.float64(self.train_y.iloc[i, 0] - y_pred) ** 2
        cost /= (2 * len(self.train_x))
        return cost

    def direction(self):
        d = [0] * 5
        for i in range(len(self.train_x)):
            error = self.train_y.iloc[i, 0] - (self.w0 + self.w1 * self.train_x.iloc[i, 0] + self.w2 * self.train_x.iloc[i, 1] + self.w3 * self.train_x.iloc[i, 2] + self.w4 * self.train_x.iloc[i, 3])
            d[0] += error
            d[1] += error * self.train_x.iloc[i, 0]
            d[2] += error * self.train_x.iloc[i, 1]
            d[3] += error * self.train_x.iloc[i, 2]
            d[4] += error * self.train_x.iloc[i, 3]
        return [-lr for lr in d]

    def update_weight(self, d):
        self.w0 -= self.learning_rate * d[0]
        self.w1 -= self.learning_rate * d[1]
        self.w2 -= self.learning_rate * d[2]
        self.w3 -= self.learning_rate * d[3]
        self.w4 -= self.learning_rate * d[4]

    def gradient_descent(self):
        for i in range(self.iter):
            cost = self.cost_find()
            d = self.direction()
            self.update_weight(d)
            self.cost_history.append(cost)
            if (i+1) % 100 == 0:  # Record cost after every 100 iterations
                print(f"Iteration: {i+1}, Cost: {cost}")
            if (i+1) % 1000 == 0:  # plot cost 
                self.plot_cost(i+1)
                self.testing()

    def plot_cost(self,i):
        print("i",i)
        iterations = range(100, i + 1, 100)
        print("itera",iterations)
        costs = self.cost_history[99::100]  # Get costs after every 100 iterations
        print("cost",costs)
        plt.plot(iterations, costs)
        plt.xlabel('Iterations')
        plt.ylabel('Cost')
        plt.title('Cost vs. Iterations')
        plt.show()
        
    def cost_test(self,i): 
        y_pred = self.w0 + self.w1 * self.test_x.iloc[i, 0] + self.w2 * self.test_x.iloc[i, 1] + self.w3 * self.test_x.iloc[i, 2] + self.w4 * self.test_x.iloc[i, 3]
        return y_pred
    
    def class_define(self,y_pred):
        if(y_pred<=1.5):
            return 1
        elif(y_pred >1.5 and y_pred <=2.5):
            return 2
        elif(y_pred>2.5):
            return 3
    
    def testing(self):
        acc=0
        for i in range(len(self.test_x)):
            y_pred=self.cost_test(i)
            y=self.class_define(y_pred)
            if(y== test_y.iloc[i,0]):
                acc=acc+1
        print("Accuracy : ", (acc/len(self.test_x))*100)
                
if __name__ == "__main__":
    # Define and initialize train_x, train_y, test_x, test_y
    obj = GradientDescent(5000, 0.0001, train_x, train_y, test_x, test_y)
    obj.gradient_descent()
```
